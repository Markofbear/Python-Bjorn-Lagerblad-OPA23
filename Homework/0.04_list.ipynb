## 0. Dice rolls (*)

Simulate 10 dice rolls and **append** the rolls to a list or use **list comprehension**.

&nbsp; a) &nbsp; sort the list in **ascending** order (*)

&nbsp; b) &nbsp; sort the list in **descending** order (*)

&nbsp; c) &nbsp; find the **maximum** and **minimum** value in the list (*)

--- 




````python
import random # importerar random från en tidigare övning

# random 1-6 10 gånger _ är för att vi inte ute efter en specifik siffra
rolls = [random.randint(1, 6) for _ in range(10)]

# a) Sort ascending
rolls_asc = sorted(rolls)

# b) descending
rolls_desc = sorted(rolls)

# c) lättare än vad jag trodde
max_roll = max(rolls)
min_roll = min(rolls)

print("All rolls:", rolls)
print("sorts asc:", rolls_asc)
print("sort dex):", rolls_desc)
print("Maximum roll:", max_roll)
print("Minimum roll:", min_roll)
````

## 1. Food menu (*)

Create a:  

&nbsp; a) &nbsp; **list** with the following elements: "vegetarisk lasagne", "spaghetti", "fisk", "grönsakssoppa", "pannkakor". (*)

&nbsp; b) &nbsp; **list** with the weekdays (*)

&nbsp; c) &nbsp; create a food menu with each day corresponding to each food item and print it out. (*)


--- 



````python
# a) Lists som i C# typ
food_items = ["vegetarisk lasagne", "spaghetti", "fisk", "grönsakssoppa", "pannkakor"]

# b) se ovan
weekdays = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday"]

# c) zip lägger ihop dom med dictionary
food_menu = dict(zip(weekdays, food_items))

# for som i C#
for day, food in food_menu.items():
    print(f"{day}: {food}")
````

## 2. Squares (*)

&nbsp; a) &nbsp; Use **list comprehension** to create a list of squares from -10 to 10 (*)

&nbsp; b) &nbsp; Plot this list using **matplotlib**. (*)

---


````python
import matplotlib.pyplot as plt # chatgpt edition(TM)

# List comprehension to create a list of squares from -10 to 10
squares = [x ** 2 for x in range(-10, 11)]
print(squares)

# Plotting the list
plt.plot(range(-10, 11), squares, marker='o')
plt.xlabel('Number')
plt.ylabel('Square')
plt.title('Squares from -10 to 10')
plt.grid(True)
plt.show()
````
