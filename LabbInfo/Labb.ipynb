```python
""""Kommer ge källkritik där det går, annars är det ifrån tidigare uppgifter,brainstorming med klasskamrater eller almäna eureka moments, kommer att ge så fördjupande kommentarer som är möjligt för att visa att jag kan och förstår vad som är skrivet"""

import matplotlib.pyplot as plt  # Importerar matplot för användning, och kallar den för plt

# Läser och rengör datan 
def process_line(line):
    parts = line.strip().replace("(", "").replace(")", "").split(",")  # Tar bort paranteser och splittar vid komma.
    return [float(coord) for coord in parts]  # Konvertera till float och ger en lista

# Läser pichu.txt
with open("../Labbinfo/pichu.txt", "r") as file:
    raw_text_pichu = file.readlines()  # kallar den för raw_text_pichu läser lines

# Se ovan ändra till "pikachu" efter smak
with open("../Labbinfo/pikachu.txt", "r") as file:
    raw_text_pikachu = file.readlines()  

pichu_data = []  # List to store processed pichu data
skip_header = True  # OBS Chatgpt lösning! Hoppar över "headern" alltså den första delen där vi har text i vår data
for line in raw_text_pichu:  # Gå över varje "line" i våran raw
    if skip_header:  
        skip_header = False  # Sätter till falsk så att programmet börjar läsa in data
        continue
    if line.strip().startswith("("):  
        pichu_data.append(process_line(line))  

# Se ovan
pikachu_data = []  
skip_header = True  
for line in raw_text_pikachu:  
    if skip_header:  
        skip_header = False  
        continue
    if line.strip().startswith("("):  
        pikachu_data.append(process_line(line))  

# Start för X och Y
pichu_widths = [item[0] for item in pichu_data]  # Extraherar "width" i pichu
pichu_heights = [item[1] for item in pichu_data]  # Extraherar "height" i pichu

pikachu_widths = [item[0] for item in pikachu_data]  
pikachu_heights = [item[1] for item in pikachu_data]  

# Plot the data for pichu and pikachu
plt.scatter(pichu_widths, pichu_heights, label="Pichu", color="Black")  # Plot pichu data
plt.scatter(pikachu_widths, pikachu_heights, label="Pikachu", color="Red")  # Plot pikachu data
plt.xlabel("Width (cm)")  # Namnet för X
plt.ylabel("Height (cm)")  # Namnet för Y
plt.title("Pichu and Pikachu Data")  # Titeln
plt.legend(loc="upper left")  # Vart vi sätter vår "legend"
plt.show()  
```
